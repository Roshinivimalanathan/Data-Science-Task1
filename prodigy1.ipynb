```python
sorted_pop_2020=tot_new_df.sort_values(by='2020',ascending=False)
sorted_pop_2020
```

```python
top_ten_count_2020=sorted_pop_2020.head(10)
top_ten_count_2020
```

```python
sorted_pop_2000=tot_new_df.sort_values(by='2000',ascending=False)
sorted_pop_2000
```

```python
top_ten_count_2000=sorted_pop_2000.head(10)
top_ten_count_2000
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))

plt.subplot(2,2,1)
plot=sns.barplot(x='2020',y="Country Name",data=top_ten_count_2020,palette="coolwarm")
plt.title("Top Ten countries of 2020")
plt.xlabel("Total Population")
plt.ylabel("Country")
#add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x='2000',y="Country Name",data=top_ten_count_2000,palette="coolwarm")
plt.title("Top Ten countries of 2000")
plt.xlabel("Total Population")
plt.ylabel("Country")
#add_value_labels(plot)

plt.tight_layout()
```
